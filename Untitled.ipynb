```python
import pandas as pd
import zipfile
import os
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Configure Tesseract (update this path if needed)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

print("Libraries imported and environment setup completed.")
```

```python
# Load annotations from CSV
detection_df = pd.read_csv(r"C:\Users\haswa.HASWANTH\Downloads\Data science project\Licplatesdetection_train.csv")
recognition_df = pd.read_csv(r"C:\Users\haswa.HASWANTH\Downloads\Data science project\Licplatesrecognition_train.csv")

# Display the first few rows of each dataset
print("Detection Data Sample:")
display(detection_df.head())

print("\nRecognition Data Sample:")
display(recognition_df.head())
```

```python
def load_images_and_boxes(df, base_path):
    images = []
    boxes = []
    for _, row in df.iterrows():
        img_path = os.path.join(base_path, row['img_id'])
        image = cv2.imread(img_path)
        images.append(image)
        boxes.append((row['xmin'], row['ymin'], row['xmax'], row['ymax']))
    return images, boxes

# Load detection images and bounding boxes
detection_images, detection_boxes = load_images_and_boxes(detection_df, r"C:\Users\haswa.HASWANTH\Downloads\Data science project\Licplatesdetection_train\license_plates_detection_train")

# Load recognition images and labels
recognition_images = [cv2.imread(os.path.join(r"C:\Users\haswa.HASWANTH\Downloads\Data science project\Licplatesrecognition_train\license_plates_recognition_train", f)) for f in recognition_df['img_id']]
recognition_labels = recognition_df['text'].tolist()

print("Loaded and preprocessed images successfully.")
```

```python
def visualize_boxes(image, box):
    """Draw bounding box on the image."""
    xmin, ymin, xmax, ymax = box
    image_with_box = image.copy()
    cv2.rectangle(image_with_box, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
    plt.imshow(cv2.cvtColor(image_with_box, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Visualize a sample detection image
visualize_boxes(detection_images[0], detection_boxes[0])
```

```python
def recognize_text(image):
    """Perform OCR on the given image using Tesseract."""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(gray, config='--psm 8')  # Single word mode
    return text.strip()

# Test OCR on a sample license plate image
sample_plate = recognition_images[0]
recognized_text = recognize_text(sample_plate)
print("Recognized Text:", recognized_text)
```

```python
def process_test_set(test_path):
    """Load test images for license plate detection and recognition."""
    test_images = []
    test_filenames = os.listdir(test_path)
    for file in test_filenames:
        img_path = os.path.join(test_path, file)
        image = cv2.imread(img_path)
        test_images.append((file, image))
    return test_images

# Load test images
test_images = process_test_set(r"C:\Users\haswa.HASWANTH\Downloads\Data science project\test\test\test")
print(f"Loaded {len(test_images)} test images.")

# Visualize a test image
plt.imshow(cv2.cvtColor(test_images[0][1], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
def detect_and_recognize(image, boxes):
    """Detect license plate and perform character recognition."""
    for box in boxes:
        xmin, ymin, xmax, ymax = box
        license_plate = image[ymin:ymax, xmin:xmax]
        text = recognize_text(license_plate)
        print(f"Detected License Plate Text: {text}")

# Test the pipeline on the first image
test_image, test_boxes = detection_images[0], [detection_boxes[0]]  # Example
detect_and_recognize(test_image, test_boxes)
```
